```python
LocustValley= pd.read_csv('LocustValley_2006_15.csv')
LocustValley.head(3)
```

```python
#LocustValley_combined['Town'] = 'Locust Valley'

generic = LocustValley

generic.rename(columns=lambda x: x.replace(' ', '_'), inplace=True)

# Combined_data.dropna(subset=['Sold_Price'], inplace=True)
generic.dropna(subset=['Sold_Price'], inplace=True)

generic =  generic[['Address', 'Town','Zip','State','Style',
                    'ML#','Sold_Price','Baths_Full', 'Baths_Half',
                    'Bath','Bedrooms',
                      'Appearance','REO' , 'School_District_Name',
                 'Approx_Int_Square_Footage','Street_Name','Contract_Date','Lot_Square_Footage','Lot_Size' ]]

# Convert 'Contract Date' to datetime
generic['Contract Date'] = pd.to_datetime(generic['Contract_Date'])

# Extract year from 'Contract Date' and create a new column
generic['Year'] = generic['Contract Date'].dt.year
unique_dates = generic['Year'].unique()

# Print unique dates
print(unique_dates)

generic.info()
# Now conver ti back to the town name
LocustValley = generic
```

```python
LocustValley.head(3)
```

```python
# Calculate the frequency of each style
style_counts = LocustValley['Style'].value_counts(normalize=True) * 100

# Identify styles that are less than 5%
styles_to_replace = style_counts[style_counts < 5].index
```

```python
# Replace styles that are less than 5% with 'Other'
LocustValley['Style'] = LocustValley['Style'].replace(styles_to_replace, 'Other')
```

```python
LocustValley.head(5)
```

```python
# Count the occurrences of each style
style_counts = LocustValley['Style'].value_counts().reset_index()
style_counts.columns = ['Style', 'Count']

# Create a pie chart
fig = px.pie(style_counts, values='Count', names='Style', title='Distribution of Styles in LocustValley')

# Show the pie chart
fig.show()
```

```python
# Define category bins and mapping
category_bins = [0, 4000, 8000, 10000, 20000, 40000, float('inf')]
bin_labels = ['40 x 100', '80 x 100', '100 x 100', '1/2 acre', '1 acre', '2 acres', '2 acres +']

# Create a function to map square footage to labels
def map_to_label(sq_ft):
    for i, bin_edge in enumerate(category_bins[:-1]):
        if sq_ft <= bin_edge:
            return bin_labels[i]
    return bin_labels[-1]

# Create a new column with labels
LocustValley['Lot_Square_Footage_Label'] = LocustValley['Lot_Square_Footage'].apply(map_to_label)

# Create the scatter plot
fig = px.scatter(LocustValley, x='Lot_Square_Footage_Label', y='Sold_Price', color='Style',
                 labels={
                     'x': 'Lot Square Footage',
                     'Sold_Price': 'Sold Price',
                     'Style': 'Style'
                 },
                 title='Lot Square Footage vs. Sold Price by Style')

# Update x-axis ticks and labels (including the last category)
fig.update_xaxes(tickvals=list(range(len(bin_labels))),  # Use all labels
                ticktext=bin_labels)

fig.show()
```
