```python
import pandas as pd

# Read the CSV file into a DataFrame
#med_incomes = pd.read_csv('Median_Incomes_9_towns.csv')
mean_incomnes =  pd.read_csv('Avg_and_Median_Incomes_9_towns - Sheet1 (1).csv')

# Define the towns to combine
roslyn_towns = ['Roslyn_Harbor', 'Roslyn', 'Roslyn_Heights']

# Create a copy of the DataFrame to avoid modifying the original
combined_data = mean_incomnes.copy()

# Combine rows with Roslyn town names
combined_data.loc[combined_data['Town'].isin(roslyn_towns), 'Town'] = 'Roslyns'

# Group data by Year and Town, calculate average household income
average_income_per_town_year = combined_data.groupby(['Year', 'Town'])['Mean_Income'].mean().reset_index()

# Update the original DataFrame with the combined data
mean_incomnes = average_income_per_town_year

# Print the updated DataFrame
print(mean_incomnes)

```

```python
# prompt: convert  Household_Income to integer

mean_incomnes['Mean_Income'] = mean_incomnes['Mean_Income'].astype(int)
```

```python
import plotly.express as px

# Make sure your DataFrame is named med_incomes and it's properly loaded

# Correct any potential mismatch in town names in the DataFrame vs. color_dict keys
# Ensure the 'Town' column of your DataFrame exactly matches these keys
color_dict = {
    'Roslyns': '#006400',  # DarkGreen
    'Port Washington': '#008000',  # Green
    'Locust Valley': '#90EE90',  # LightGreen
    'Mineola': '#00008B',  # DarkBlue
    'Westbury': '#0000FF',  # Blue
    'Farmingdale': '#ADD8E6',  # LightBlue
    'Freeport': '#8B0000',  # DarkRed
    'Hempstead': '#FF0000',  # Red
    'Roosevelt': '#F08080'  # LightCoral
}

# Specify the order of towns
town_order = ['Roosevelt', 'Hempstead', 'Freeport', 'Farmingdale', 'Mineola',
              'Westbury', 'Roslyns', 'Port Washington', 'Locust Valley']

# Use the 'category_orders' parameter to specify the order of towns
fig = px.bar(mean_incomnes,
             x="Year",
             y="Mean_Income",
             color="Town",
             barmode='group',
             color_discrete_map=color_dict,
             category_orders={"Town": town_order})

fig.update_layout(
    title="Average Household Income by Town and Year",
    xaxis_title="Year",
    yaxis_title="Average Household Income",
    legend_title="Town"
)

# This ensures the legend order matches the town order specified
fig.update_traces(marker=dict(line=dict(width=0)))
fig.update_layout(legend=dict(traceorder="normal"))

fig.show()
```
