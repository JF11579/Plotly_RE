```python
# Here

import plotly.express as px
import pandas as pd

# Assuming 'data' is your DataFrame
# (Your existing data code should go here)

# Median Category data
median_category_data = pd.DataFrame({
    'Town': ['New Hyde Park', 'Bethpage', 'Levitown', 'Mineola', 'Great Neck', 'Roslyns',
             'Port Washington', 'East Meadow', 'Westbury', 'Hicksville', 'Freeport',
             'Hempstead', 'Uniondale', 'Roosevelt', 'Baldwin', 'Floral Park',
             'East Rockaway', 'Farmingdale', 'Locust Valley', 'Oyster Bay Cove', 'Sands Point'],
    'Median_Sold': [800000, 650000, 612000, 700000, 1400000, 1300000, 1100000, 710000,
                    715000, 659000, 565000, 560000, 582000, 554000, 608000, 765000,
                    579000, 680000, 950000, 2000000, 2446000],
    'Median_Category': ['Above Median', 'Above Median', 'Median', 'Median', 'Below Median',
                        'Below Median', 'Below Median', 'Below Median', 'Below Median',
                        'Below Median', 'Above Median', 'Below Median', 'Median',
                        'Above Median', 'Above Median', 'Above Median']
})

# Merge the median category data with your sorted data
sorted_data = median_category_data.sort_values(by='Median_Sold')

# Map Median_Category to colors
color_map = {'Above Median': 'green', 'Median': 'blue', 'Below Median': 'red'}

# Create the bar chart with the sorted and color-mapped data
bars = px.bar(sorted_data,
              x='Town',
              y='Median_Sold',
              color='Median_Category',
              color_discrete_map=color_map,
              labels={'Median_Category': 'Category'})

bars.update_layout(title='Median Sold Price Nassau County, NY 2023',
                   xaxis_title='Town',
                   yaxis_title='Median Sold Price',
                   legend_title='Category')

# Add a horizontal line at y=710000
bars.add_hline(y=710000, line_dash="solid", line_color="black", line_width=4)

# Define the start and end points of the towns on the x-axis
start_x = -0.5  # Starting just before the first bar
end_x = len(sorted_data['Town']) - 0.5  # Ending just after the last bar

# Add a shaded area for the range ±50000 from 710000
bars.add_shape(type="rect",
               x0=start_x, y0=660000, x1=end_x, y1=760000,
               line=dict(width=0),
               fillcolor="gray", opacity=0.5)

bars.show()
```
