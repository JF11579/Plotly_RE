## Were some developmets particuarly prone to not coming up with Lat & Long?

```python


# Assuming 'df' is your DataFrame after geocoding

# Group by 'DEVELOPMENT' and calculate the percentage of missing coordinates
development_percentage_missing = condos.groupby('Development').apply(lambda group: (group['latitude'].isna() | group['longitude'].isna()).mean() * 100)

print(development_percentage_missing)
```

```python
import pandas as pd

# Convert the Series to a DataFrame
Pct_missing_df = pd.DataFrame({'Percentage_Missing': development_percentage_missing})

print(Pct_missing_df)
```

```python
Pct_missing_df.shape
```

```python
# Assuming 'df' is your DataFrame
column_names = Pct_missing_df.columns.tolist()

print(column_names)
```

```python
# Move existing index to a new column


df = Pct_missing_df
# Store the original index column
original_index_column = df.index

# Reset the index
df.reset_index(drop=True, inplace=True)

# Add the original index column back as a new column
df['Original_Index'] = original_index_column

print(df)
```

```python
# Filter rows where 'Percentage_Missing' is greater than 34
filtered_df = df[df['Percentage_Missing'] > 34]

print(filtered_df)
```
