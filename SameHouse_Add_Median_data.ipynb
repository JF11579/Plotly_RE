# Merge the 2 DFs on YEAR

```python
# prompt: how to merge these 2 dataframes on YEAR?

merged_df = pd.merge(data, Fed_Median_Nassau, on='Year')
```

```python
merged_df.head(5)
```

```python
merged_df.info()
```

```python
# Convert non-numeric strings to NaN
merged_df['MedianPriceIndex'] = pd.to_numeric(merged_df['MedianPriceIndex'], errors='coerce')

# Fill NaN values with a default integer (e.g., 0)
merged_df['MedianPriceIndex'] = merged_df['MedianPriceIndex'].fillna(399)

# Now convert to integer
merged_df['MedianPriceIndex'] = merged_df['MedianPriceIndex'].astype(int)
```

```python
merged_df.head(5)
```

### A word about the *MedianPrice* .  This is an index number where the January 1995 value = 100.  So when the 2006 *MedianPrice* index = 186 thant meand the home has appreciated 86% since January 1995.

### Not sure why that is better.  More durable, perhaps.  But we will create a new column where we convert that Index number intoa dollar amound for easier reading.

Zillow says the 2023 median price in nassau was $710.  good benchmark to check.

But these numbers come from the St Louis Fred and the FED only tracks Conforming loans So the actual median would be higher.

[NYT article](https://www.nytimes.com/1997/03/23/realestate/for-region-s-housing-a-solid-96.html) states that the median sold price in nassau in 1995 = $168k




```python
# prompt: adda column called 'BasePrice' and populate it 168000

merged_df['BasePrice'] = 168000
```

```python
merged_df.head(5)
```

```python
# prompt: Create new column called 'MedianPriceDecimal' by multiplying MedianPrice collumn by 0.01

merged_df['MedianPriceDecimal'] = merged_df['MedianPriceIndex'] * 0.01
```

```python
merged_df.head(2)
```

```python
# prompt: create new column called 'MedianPrice_DollarCoversion' where we multiply BasePrice by  MedianPriceDecimal

merged_df['MedianPrice_DollarConversion'] = merged_df['BasePrice'] * merged_df['MedianPriceDecimal']
```

```python
merged_df.head(2)
```

```python
merged_df.tail(20)
```

```python
# prompt: now convert MedianPrice_DollarConversion to integers

merged_df['MedianPrice_DollarConversion'] = merged_df['MedianPrice_DollarConversion'].astype(int)
```

# Test Plot

```python
import pandas as pd
import plotly.express as px

# Assuming data is your DataFrame with 'Year', 'Town', and 'Sold_Price' columns

# Calculate the median of Sold_Price for each year and town
median_prices = merged_df.groupby(['Year', 'Town'])['Sold_Price'].median().reset_index()
#
# Convert 'Year' to string to ensure Plotly treats it as a categorical variable
median_prices['Year'] = median_prices['Year'].astype(str)

# Create a multi-line chart using Plotly with a line for each town
fig = px.line(median_prices, x='Year', y='Sold_Price', color='Town',
              title='Median Sold Prices by Town Over Years')
fig.show()
```

# Add new line for medain price

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

# Assuming data is your DataFrame with 'Year', 'Town', 'Sold_Price', and 'MedianPrice_DollarConversion' columns

# Calculate the median of Sold_Price for each year and town
median_prices = merged_df.groupby(['Year', 'Town'])['Sold_Price'].median().reset_index()

# Convert 'Year' to string to ensure Plotly treats it as a categorical variable
median_prices['Year'] = median_prices['Year'].astype(str)

# Create a multi-line chart using Plotly with a line for each town
fig = px.line(median_prices, x='Year', y='Sold_Price', color='Town',
              title='Median Sold Prices by Town Over Years')

# Add a line for MedianPrice_DollarConversion
median_conversion = merged_df.groupby('Year')['MedianPrice_DollarConversion'].median().reset_index()
median_conversion['Year'] = median_conversion['Year'].astype(str)

fig.add_trace(go.Scatter(x=median_conversion['Year'], y=median_conversion['MedianPrice_DollarConversion'],
                         mode='lines', name='Conversion Rate',
                         line=dict(color='black', width= 6)))

fig.show()
```
