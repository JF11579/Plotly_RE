```python
# test
# Add a small random noise to the x-values for jitter effect
# Convert 'DEVELOPMENT' column to strings and add jitter
jitter_amount = 0.01  # Adjust this value for desired jitter effect
condos['DEVELOPMENT_jittered'] = (condos['Development'].astype(str) +
                               np.random.uniform(-jitter_amount, jitter_amount, len(condos)).astype(str))


# Create a scatter plot
fig = px.scatter(condos, x='Development', y='Sold_Price', color='Town')

# Define custom hover template
hover_template = "<b>ML#</b>: %{customdata[0]}<br><b>Address</b>: %{customdata[1]}<br><b>Development</b>: %{customdata[2]}<br><b>Sold Price</b>: $%{y}"

# Update trace with custom hover template
fig.update_traces(customdata=condos[['ML#', 'full_address', 'Development']], hovertemplate=hover_template)

# Set custom y-axis tick format
fig.update_yaxes(tickformat='$')
# Show the plot
fig.show()
#
# Save the plot as an interactive HTML file
pio.write_html(fig, 'Condo_sold_scatterplot_9_21_23_1.html')
```
